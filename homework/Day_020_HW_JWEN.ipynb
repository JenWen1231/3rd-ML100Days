```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
%matplotlib inline
plt.style.use('ggplot')
warnings.filterwarnings('ignore')
```

```python
matrix=np.random.normal(size=(10,10))
plt.figure(figsize=(10,10))
sns.heatmap(matrix,cmap=plt.cm.RdYlBu_r,vmin=-0.1,annot=True,vmax=1)
plt.title('Random Heatmap')
#heatmap=??
plt.show()
```

```python
nrow=1000
ncol=3

matrix=np.random.normal(size=(1000,3))
indice=np.random.choice([0,1,2],size=nrow)
plot_data=pd.DataFrame(matrix,indice).reset_index()

grid=sns.PairGrid(data=plot_data, size=3, diag_sharey=False, hue='index',vars=[x for x in list(plot_data.columns) if x!='index'])

grid.map_upper(plt.scatter,alpha=0.2)
grid.map_diag(sns.distplot(plot_data,bins=100,color='k'))
grid.map_lower(sns.kdeplot,cmap=plt.cm.OrRd_r)
plt.suptitle('Three chars',size=32,y=1.05)
plt.show()
```

```python
nrow=1000
ncol=3
matrix=np.random.normal(size=(1000,3))
indice=np.random.choice([0,1,2],size=nrow)
plot_data=pd.DataFrame(matrix,indice).reset_index()

grid=sns.PairGrid(data=plot_data,size=3,diag_sharey=False,hue='index',vars=[x for x in list(plot_data.columns) if x!='index'])

grid.map_upper(plt.scatter,alpha=0.2)
grid.map_diag(plot_data.plot.bar())
grid.map_lower(sns.kdeplot,cmap=plt.cm.OrRd_r)

plt.show()
```
