```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use('ggplot')

import warnings
warnings.filterwarnings('ignore')

dir_data='./Data/'
f_app=os.path.join(dir_data,'application_train.csv')
app_train=pd.read_csv(f_app)
app_train.head()
```

```python
app_train['DAYS_BIRTH']=abs(app_train['DAYS_BIRTH'])
app_train['DAYS_BIRTH']
```

```python
age_data=app_train[['TARGET','DAYS_BIRTH']]
age_data
```

```python
age_data['YEARS_BIRTH']=age_data['DAYS_BIRTH']/365
age_data
```

```python
age_cut=[20,25,30,35,40,45,50,55,60,65,70]

age_data['AGE_GROUP']=pd.cut(age_data['YEARS_BIRTH'].values,age_cut,include_lowest=True)
bin_cut=len(age_data['AGE_GROUP'].value_counts())
age_data['AGE_GROUP']
```

```python
print(age_data['AGE_GROUP'].value_counts())

print(len(age_data['AGE_GROUP'].value_counts()))
```

```python
age_data['YEARS_BINNED']=pd.cut(age_data['YEARS_BIRTH'],bins=bin_cut)
print(age_data['YEARS_BIRTH'].value_counts())
```

```python
age_data.head()
```

```python
age_data.shape
```

```python
year_group_sorted=age_data['YEARS_BINNED'].sort_values()
year_group_sorted
```

```python
len(year_group_sorted)
```

```python
plt.figure(figsize=(8,6))
for i in range(len(year_group_sorted)):
    sns.distplot(age_data.loc[(age_data['YEARS_BIRTH']==year_group_sorted[i]) & \
                 age_data['TARGET']==0,'YEARS_BIRTH'],label=str(year_group_sorted[i]))
plt.title('KDE with Age groups')
plt.show()
```

```python
age_groups=age_data.groupby('YEARS_BINNED').mean()
age_groups
```

```python
plt.figure(figsize=(8,8))

px=age_data['YEARS_BINNED']
py=app_train['TARGET']
sns.barplot(px,py)

plt.xticks(rotation=75);plt.xlable('Age Group (years)');plt.ylabel('Failure to Repay (%)')
plt.title('Failure to Repay by Age Group');
plt.show()
```
